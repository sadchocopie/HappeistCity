```python
#Cost of living index (past 18 month)
#ranked from high to low
#https://www.numbeo.com/cost-of-living/country_result.jsp?country=United+States
#contains all cities from above website but not 

import numpy as np
import pandas as pd

import requests
from bs4 import BeautifulSoup
page = requests.get("https://www.numbeo.com/cost-of-living/country_result.jsp?country=United+States")
```

```python
soup = BeautifulSoup(page.content, 'html.parser')
table = soup.find(id="t2")
citiesTable = table.find_all(class_="discreet_link")
city = citiesTable[0]
print(city.prettify())
```

```python
city = city.get_text()
print(city)
```

```python
city_list= []
for x in citiesTable:
    x.prettify()
    x = x.get_text()
    city_list.append(x)

print(city_list)
```

```python
table = soup.find(id="t2")
indicesTable = table.find_all("td")

print(indicesTable)

index_list =[]

# add each remaining index elemnts to list
count = 0
while (count <= 132):
    index_list.append(indicesTable[2 + (8 * count)].get_text())
    count = count+1
print(index_list)
```

```python
city_index_df = pd.DataFrame(
    {'city': city_list,
     'index': index_list
    })

city_index_df
```
